```python
def distance(initial , seconder): 
    d = ((initial[0] - seconder[0])**2  + (initial[1] - seconder[1])**2)*0.5 
    return d
```

```python
filtre = (array > 200)
a = np.transpose(np.where(filtre))
responsability_list = [a[0,:]]

commit_reject = []
counter = 0


for i in a:
    d = distance(responsability_list[counter] , i)
    print(d)
```
