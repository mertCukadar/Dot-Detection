```python
import cv2
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from PIL import Image
```

```python

img = Image.open("test2.png" , mode = "r", formats=None)
img = np.array(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage)
```

```python
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 105, 100, cv2.THRESH_BINARY)
```

```python
plt.imshow(blackAndWhiteImage)
```
