```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = cv2.imread("hyttest2.png")
gray_img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
# get inverse intesity 
gray_array = np.array(gray_img)
inverse_intensity = np.absolute(gray_array - (255/1))
```

```python
# lets subploting all of them 
plt.subplot(1,2,1)
plt.title("gray")
plt.imshow(gray_img)
plt.subplot(1,2,2)
plt.title("inverse gray")
plt.imshow(inverse_intensity)
plt.show()
```

let us collect the region respec to speciffied range list

```python
x_range = np.arange(0,2593,162)
y_range = np.arange(0,1945,162)

x_range = np.append(x_range , 0)
y_range = np.append(y_range , 0)



print(f"x range : {x_range}")
print(f"y range : {y_range}")
print(f"shape : {np.shape(inverse_intensity)}")
```

```python
thresholder = (inverse_intensity > 150)
Black = (inverse_intensity < 150)
```

```python
for j in range(len(y_range)):
    for i in range(len(x_range)):
        local_region = inverse_intensity[x_range[i]: x_range[i+1] , y_range[j] : y_range[j+1]]
        thresholder = (local_region > 150)
        Black = (local_region < 150)
        local_region[thresholder] = 255
        local_region[Black] = 0
        inverse_intensity[x_range[i]: x_range[i+1] , y_range[j] : y_range[j+1]] = local_region
        if i == 11:
            break
    if j == 9:
        break
    
    

 
      
```
