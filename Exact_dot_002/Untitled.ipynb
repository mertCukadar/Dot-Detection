```python
from PIL import Image , ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
import cv2





img = Image.open('inverse_image.png')

plt.imshow(img)
plt.show()
```

```python
enhancer = ImageEnhance.Contrast(img)

```

```python
factor = 0.5 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('original-image.png')

factor = 2 #decrease constrast
im_output = enhancer.enhance(factor)
im_output.save('less-contrast-image.png')

factor = 0.3

#increase contrast
im_output = enhancer.enhance(factor)
im_output.save('more-contrast-image.png')
```

```python
img2 = Image.open("more-contrast-image.png")

enhancer = ImageEnhance.Brightness(img2)

factor = 0.5 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('original-image.png')

factor = 1 #darkens the image
im_output = enhancer.enhance(factor)
im_output.save('darkened-image.png')

factor = 0.3 #brightens the image
im_output = enhancer.enhance(factor)
im_output.save('brightened-image.png')
```
